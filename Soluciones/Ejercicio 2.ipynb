# Hamlet
Escriba un c√≥digo que lea el archivo `hamlet.txt` y encuentre las n_palabras palabras de mas de `n_letras` que mas se repiten en el texto. Por ejemplo, si `n_palabras=5` y `n_letras=3` el codigo debe encontrar las cinco palabras de mas de tres letras que mas se repiten en el texto.

```python
n_palabras=5
n_letras=3
```

```python
with open("hamlet.txt", "r") as hamlet_file:
    text = hamlet_file.read() # cargar el texto
print(text)
```

```python
clean_text = text.replace('\n', ' ') # delimitador de linea
clean_text = clean_text.lower() # minuscula
print(clean_text)
```

```python
# posibles signos de puntuacion
puntuaciones = [',', '.', ';', ':', '?', '!', '*', "'", '"']

for puntuacion in puntuaciones:
    clean_text = clean_text.replace(puntuacion, '')
```

```python
todas_palabras = clean_text.split(' ') # separa por espacios
print(todas_palabras)
```

```python
palabras_interes = {} # diccionario con palabras de longitud >= n_letras
for palabra in todas_palabras:
    if len(palabra) >= n_letras:
        if palabra in palabras_interes:
            palabras_interes[palabra] += 1
        else:
            palabras_interes[palabra] = 1
            
print(palabras_interes)
```

```python
palabras_repeticion = sorted(palabras_interes, key=palabras_interes.get,
                                                reverse=True)[:n_palabras] # organiza y toma las primeras n_palabras

for palabra in palabras_repeticion:
    print(palabra, palabras_interes[palabra])
```
