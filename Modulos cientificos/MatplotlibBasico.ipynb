```python
# Los marcadores pueden tener diferentes formas
plt.figure()
markers = ["o","^","s","+","x","D"]
# o -> circulo
# ^ -> triángulo
# s -> cuadrado
# + -> plus
# x -> cruz
# D -> diamante
k = 0
for i in markers:
    puntos = [k*j for j in range(-10,11)]
    abscisas = range(-10,11)
    plt.plot(abscisas,puntos,i) # Solo los puntos
    plt.plot(abscisas,puntos) # Los puntos unidos con líneas
    k=k+1
plt.show()
```

```python
# También las líneas pueden tener diferentes formatos
plt.figure()
tiposl = ["-","--","-.",":"]
k = 1
for i in tiposl:
    x = np.linspace(0,1,10)
    y = k * np.ones(10)
    plt.plot(x,y,i,lw=10)
    k = k + 1
plt.ylim(0,5) # Para modificar el rango vertical mostrado
plt.show()
```

```python
# Rosa polar
# http://en.wikipedia.org/wiki/Rose_%28mathematics%29
figure(figsize=(10,10))
k = 1
for n in range(1,8):
    for d in range(1,8):
        subplot(7,7,k)
        kah = float(n)/d
        angulos = linspace(0,2*d*pi,200)
        radios = sin(kah * angulos)
        x = radios * cos(angulos)
        y = radios * sin(angulos)
        plot(x,y)
        axis("off") # esto elimina los ejes
        axis("equal")
        k = k + 1
show()
```

## Leyenda

```python
# Cuando se ponen varias gráficas en un mismo eje
# conviene poner una leyenda que especifique cual es cual
figure()
x = linspace(0,2*pi,20)
plot(x,sin(x),"or-",label='sin(x)')
plot(x,cos(x),"ob-",label='cos(x)')
legend()
axis([0,2*pi,-1,1])
show()
```

```python
# La posición de la leyende puede modificarse con la ayuda
# del parámetro loc
locs = ['best', 'upper right','upper left','lower left','lower right',
        'right','center left','lower center','upper center', 'center']
figure(figsize=(15,20)) # fijarse en la opción para cambiar el tamaño de la gráfica, unidad es pulgadas
k = 1
x = linspace(0,2*pi,100)
y = sin(x)
for i in locs:
    subplot(5,2,k)
    plot(x,sin(x),label='$\sin(x)$')
    plot(x,cos(x),label='$\cos(x)$')
    legend(loc=i)
    title(i)
    k = k + 1
    xlim(0,2*pi)
subplots_adjust(hspace=.5)
show()
```

## Imshow

```python
#####################
# Impresora Radical #
#####################

n=7 # Este es el lado de la matriz cuadrada
# Se inicializa toda la matriz con ceros
mat=[[0 for i in range(n)] for j in range(n)]
# Lo siguiente para comenzar en la esquina superior izquierda
i=0
j=0
# Cuatro modos: a la derecha (0), abajo (1), izquierda (2) y arriba (3).
# Si se "choca" contra el borde de la matriz o con un elemento ya 
# relleno entonces se cambia de modo.
# Comenzar moviéndose hacia la derecha.
mode=0
for k in range(1,n*n+1):
    # Rellenar el elemento actual con el número correspondiente
    mat[i][j]=k
    # Decidir el modo
    if mode == 0:
        if j==n-1 or mat[i][j+1]>0:
            mode=1
    elif mode == 1:
        if i==n-1 or mat[i+1][j]>0:
            mode=2
    elif mode == 2:
        if j==0 or mat[i][j-1]>0:
            mode=3
    elif mode == 3:
        if i==0 or mat[i-1][j]>0:
            mode=0
    # Actuar de acuerdo al modo
    if mode == 0:
        j+=1
    elif mode == 1:
        i+=1
    elif mode == 2:
        j-=1
    elif mode == 3:
        i-=1
```

```python
figure(figsize=(10,5))
subplot(1,2,1)
imshow(mat)
title(u"Con interpolación")
subplot(1,2,2)
imshow(mat,interpolation="None")
title(u"Sin interpolación")
show()
```
